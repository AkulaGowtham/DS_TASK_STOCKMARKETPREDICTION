```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
numdf = []
directory = "D:/Data Science/Stock Market/SMA DataSets"
for file in os.listdir(directory):
    if file.endswith('.csv'):
        df = pd.read_csv(os.path.join(directory, file))
        numdf.append(df)
```

```python
combined_df1 = pd.concat(numdf, ignore_index=True)
columns_to_drop = ['Company Name', 'Industry', 'ISIN Code']
combined_df1.drop(columns=columns_to_drop, inplace=True)
combined_df1
```

```python
metapath = "D:/Data Science/Stock Market/SMA DataSets/stock_metadata.csv"
metadf = pd.read_csv(metapath)
metadf
```

```python
nifty = pd.merge(combined_df1, metadf, on = ['Symbol', 'Series'], how = 'inner')
nifty.head()
```

```python
nifty.info()
```

```python
# Check null values
nifty.isnull().sum()
```

```python
nifty['Date'] = pd.to_datetime(nifty['Date'], errors='coerce')
nifty['Date'] = nifty['Date'].ffill()
```

```python
# Handling Null values
nifty.fillna(value = 0.0, inplace = True)
nifty.isnull().sum()
```

```python
nifty.columns
```

```python
nifty.describe()
```

```python
#Converting categorical to numerical
nifty['Date'] = pd.to_datetime(nifty['Date'])
print(nifty['Date'])
```

```python
y_price = nifty['Close']
y_boom = nifty['Open'] # Assuming 'Close' price is the target variable
x = nifty[['High', 'Low', 'Prev Close', 'Volume', 'Turnover', 'Trades', 'Deliverable Volume', '%Deliverble']]
print("Shape of X:", x.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train_price, y_test_price,y_train_boom,y_test_boom = train_test_split(x, y_price,y_boom, test_size=0.3, random_state=42)
```

```python
from sklearn import preprocessing
train1_scaler = preprocessing.StandardScaler().fit(X_train)
X_train_scaled = train1_scaler.transform(X_train) 

test_scaler = preprocessing.StandardScaler().fit(X_test)
X_test_scaled = test_scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score

price_model = LinearRegression()
price_model.fit(X_train_scaled, y_train_price)
y_pred_price = price_model.predict(X_test_scaled)
mse = mean_squared_error(y_test_price, y_pred_price)
R2_score = r2_score(y_test_price,y_pred_price)
print(mse)
print(R2_score)
```

```python

boom_model = LinearRegression()
boom_model.fit(X_train_scaled,y_train_boom)
y_pred_boom = boom_model.predict(X_test_scaled)
mse = mean_squared_error(y_test_boom,y_pred_boom)
print("MSE : ",mse)
print(r2_score(y_test_boom,y_pred_boom))
```

```python
price_forecast = price_model.predict(X_train_scaled)
boom_predictions = boom_model.predict(X_train_scaled)
```

```python
stocks = x
for stock, price, boom in zip(stocks, price_forecast, boom_predictions):
    if boom == 1:
        print(f"Stock {stock} is likely to experience a significant price increase. Forecasted closing price: {price}")
    else:
        print(f"Stock {stock} is not expected to experience a significant price increase. Forecasted closing price: {price}")
```

```python
import matplotlib.pyplot as plt

significant_increase_stocks = []
significant_increase_prices = []
no_significant_increase_stocks = []
no_significant_increase_prices = []

# Separate stocks based on boom predictions
for stock, price, boom in zip(stocks, price_forecast, boom_predictions):
    if boom == 1:
        significant_increase_stocks.append(stock)
        significant_increase_prices.append(price)
    else:
        no_significant_increase_stocks.append(stock)
        no_significant_increase_prices.append(price)

# Create area chart
plt.figure(figsize=(12, 6))

# Plot significant price increase
plt.fill_between(significant_increase_stocks, significant_increase_prices, color='red', alpha=0.5, label='Significant Price Increase')

# Plot no significant price increase
plt.fill_between(no_significant_increase_stocks, no_significant_increase_prices, color='blue', alpha=0.5, label='No Significant Price Increase')

plt.xlabel('Stock')
plt.ylabel('Forecasted Closing Price')
plt.title('Forecasted Closing Prices by Stock')
plt.legend()
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)
plt.tight_layout()
plt.show()
```
